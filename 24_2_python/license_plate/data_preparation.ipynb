```python
import os 
import shutil
from glob import glob 
from sklearn.model_selection import train_test_split
```

```python
images_dir = './data_images'
output_dir = 'data_images'
```

```python
txt_files = glob(f'{images_dir}/*.txt')
```

```python
txt_files
```

```python
data = []
```

#### find a corresponding image from the txt file

```python
for txt_file in txt_files:
  basename = os.path.splitext(os.path.basename(txt_file))[0]

  image_file = None
  for ext in ['.jpg', '.jpeg']:
    possible_image = f'{images_dir}/{basename}{ext}'
    if os.path.exists(possible_image):
      image_file = possible_image
      break 

  if image_file:
    data.append((image_file, txt_file))
print(data)
```

```python
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
test_data
```

```python
def move_files(files, destination):
  for image_file, txt_file in files:
    # moving the image file
    shutil.move(image_file, destination)
    # moving the txt file
    shutil.move(txt_file, destination)
```

```python
move_files(train_data, 'data_images/train')
```

```python
move_files(test_data, 'data_images/test')
```
